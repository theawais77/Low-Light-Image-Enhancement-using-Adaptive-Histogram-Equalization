# Low-Light Image Enhancement Experiments

Run this notebook after generating or downloading images. It uses the same project code as the command-line runner.

````python
from pathlib import Path
from src.enhancement import EnhancementParams, get_methods
from src.image_io import read_rgb, make_contact_sheet
from src.metrics import full_metrics

params = EnhancementParams()
image_path = Path('data/sample_low_light/sample_01.png')
reference_path = Path('data/sample_reference/sample_01.png')
original = read_rgb(image_path)
reference = read_rgb(reference_path)
outputs = [('original', original)]
for name, method in get_methods().items():
    enhanced = method(original, params)
    outputs.append((name, enhanced))
    print(name, full_metrics(enhanced, original, reference))

make_contact_sheet(outputs, tile_width=260)
````
